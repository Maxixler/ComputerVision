```python
import cv2
```

```python
image = cv2.imread('road(1).jpg')
cv2.imshow("image",image)
h, w = image.shape[:2]
print("Height = {}, Width = {}".format(h, w))
(B, G, R) = image[100, 100]
print("R = {}, G = {}, B = {}".format(R, G, B))
```

```python
B = image[100, 100, 0]
print("B = {}".format(B))
```

```python
roi = image[100 : 500, 200 : 700]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
```
